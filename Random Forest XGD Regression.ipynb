```python
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDRegressor, SGDClassifier
from sklearn.model_selection import train_test_split as split, KFold, StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score

```

```python
from sklearn.datasets import load_breast_cancer
bc_data = load_breast_cancer()
bc = pd.DataFrame(bc_data.data, columns=bc_data.feature_names)
bc['target'] = bc_data.target
```

```python
bc.head()
```

```python
train_bc, test_bc = split(bc, test_size=.20,random_state=2022)
```

```python
sc = StandardScaler()
train_bc_xs = sc.fit_transform(train_bc.drop(columns='target'))
test_bc_xs = sc.transform(test_bc.drop(columns='target'))
```

```python
sgd_classs = SGDClassifier()
```

```python
penalty = ['l2','l1','elasticnet']
alpha = [1,10,100,1000]
l1_ratio = [0,1]
learning_rate = ['constant','adaptive']
eta0 = [.01,.001,.2,.3,.03]
param_dist = dict(penalty=penalty,alpha=alpha,l1_ratio=l1_ratio,learning_rate=learning_rate,eta0=eta0)
```

```python
param_dist
```

```python
grid_class = GridSearchCV(sgd_classs,param_grid = param_dist, n_jobs=-1,cv=5)
```

```python
grid_class.fit(train_bc_xs,train_bc.target.values)
```

```python
grid_class.best_estimator_
#l1_ratio=Mixing parameter
```

```python
#generally you no need to do hyper tuning becuase the gridsercah CV has an amazing option
#that has option of refit is True means uses best parameters
#verbose is to give feedback on the steps
```

```python
y_pred=grid_class.predict(test_bc_xs)
```

```python
random_class = RandomizedSearchCV(sgd_classs,param_distributions = param_dist, n_jobs=-1,cv=5, verbose=5, n_iter=100)
```

```python
random_class.fit(train_bc_xs,train_bc.target.values)
```

```python
y_pred=random_class.predict(test_bc_xs)
```

```python
pd.crosstab(y_pred,test_bc.target)
```

```python
def gini_cal(n1, n2):
    return (1-pow(n1/(n1+n2),2)-pow(n2/(n1+n2),2))
```

min_sample_leaf - The minimum number of samples required to be at a leaf node.
min_sample_split - The minimum number of samples required to split an internal node



```python
dt_prune = DecisionTreeClassifier(min_samples_split=45, min_samples_leaf=15)
```

```python
dt_prune.fit(train_bc_xs,train_bc.target.values)
```

```python
fig = plt.figure(figsize=(120,50))
treeplot = tree.plot_tree(dt_prune, feature_names=list(train_bc.drop(columns='target').columns),filled=True, class_names=['Malignant','Banign'])
plt.savefig('dt.png')
```

```python
# Significant thumb rule : 3 % of rows in leafs i.e min_sample_split
# 3times of min_sample_leaf =min_samples_leaf
# we have to pass both together
```
