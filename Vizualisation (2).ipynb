```python
## Find the relationship between Survived and Age 


## Create Age Buckets 

titanic['Age_Bucket']=0
titanic.loc[titanic['Age']<16,'Age_Bucket']='0'
titanic.loc[(titanic['Age']>=16) & (titanic['Age']<32),'Age_Bucket']='1'
titanic.loc[(titanic['Age']>=32) & (titanic['Age']<48),'Age_Bucket']='2'
titanic.loc[(titanic['Age']>=48) & (titanic['Age']<64),'Age_Bucket']='3'
titanic.loc[(titanic['Age']>=64),'Age_Bucket']='4'

```

```python
titanic[['Age','Age_Bucket']]
```

```python
sns.factorplot('Age_Bucket','Survived',data=titanic,hue='Sex')
```

```python
sns.factorplot('Age_Bucket','Survived',data=titanic,col='Pclass',hue='Sex')
```
