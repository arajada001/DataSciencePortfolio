```python
## these all NA's has meaning of No basement and no fireplace i am assigninng with category of string 'NA'
```

```python
Obj_col=train.select_dtypes(include='object').columns
train[Obj_col]=train[Obj_col].fillna('No_Basement_fireplace')
```

```python
train.duplicated().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Get categorical columns
cat_cols = train.select_dtypes(include='object').columns

# Initialize OneHotEncoder
encoder = OneHotEncoder(drop='first',sparse_output=False)

# Fit and transform the categorical columns
encoded_array = encoder.fit_transform(train[cat_cols])

# Get feature names for the encoded columns
encoded_feature_names = encoder.get_feature_names_out(cat_cols)

# Create DataFrame from encoded array
encoded_df = pd.DataFrame(encoded_array, columns=encoded_feature_names, index=train.index)

# Drop original categorical columns and add encoded ones to main dataset
df = train.drop(columns=cat_cols,inplace=True)
df = pd.concat([train, encoded_df], axis=1)
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pca=PCA()
sc=StandardScaler()
```
